```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("C:\\Users\\suzyc\\Desktop\\School\\Spring 2024\\Data Mining\\HRDataset_v14.csv")
df.head()
```

```python
df.shape 
```

```python
df = df[["MaritalStatusID", "MaritalDesc", "GenderID", "Salary"]]
df = df.sort_values(by=["MaritalStatusID"])
df.head()
```

```python
male_df = df[df["GenderID"]  == 1]
male_df
```

```python
female_df = df[df["GenderID"] == 0] 
female_df
```

```python
male_df = male_df.drop("GenderID", axis = 1)
female_df = female_df.drop("GenderID", axis = 1)
```

```python
women_plot = sns.barplot(data = female_df, x = "MaritalDesc", y = "Salary")
plt.title("Marital Status & Salary for Women")
plt.ylabel("Salary")
plt.xlabel("Marital Status")
plt.savefig("WomenPlot.png")
```

```python
sns.barplot(data = male_df, x = "MaritalDesc", y = "Salary")
plt.title("Marital Status & Salary for Men")
plt.ylabel("Salary")
plt.xlabel("Marital Status")
plt.savefig("MenPlot.png")
```

```python
#sns.barplot(data = df, x = "Salary", y = "MaritalDesc", hue = "GenderID"

labels=["Women","Men"]

plt.figure(figsize=(10,8))
gender_plot  = sns.barplot(data = df, x = "Salary", y = "MaritalDesc", hue = "GenderID")
gender_plot.set_ylabel("Marital Status")
gender_plot.set_title("Salary for Marital Statuses and Genders")

h, l = gender_plot.get_legend_handles_labels()
gender_plot.legend(h, labels, title="Gender")
plt.savefig("GenderPlot.png")
```
